# IMPORT LIBRARY

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import LabelEncoder
```

# LOAD DATASET

```python
from google.colab import files

# upload a file
uploaded = files.upload()

# read the uploaded CSV file with proper encoding
for filename in uploaded.keys():
    df = pd.read_csv(filename, encoding="utf-8", keep_default_na=False)

# display the first 5 rows of the DataFrame
df.head()
```

# FEATURE EXTRACTION

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# feature extraction using TF-IDF
X = df['text_akhir']  # preprocessed text
y = df['sentiment']   # sentiment labels

# initialize TF-IDF Vectorizer
tfidf = TfidfVectorizer(
    min_df=5,          # ignore terms that appear in fewer than 5 documents
    max_df=0.8,        # ignore terms that appear in more than 80% of documents
    sublinear_tf=True, # apply sublinear term frequency scaling
    use_idf=True       # enable inverse document frequency weighting
)

# transform the text data into TF-IDF features
X_tfidf = tfidf.fit_transform(X)

# convert the TF-IDF result to a DataFrame
features_df = pd.DataFrame(X_tfidf.toarray(), columns=tfidf.get_feature_names_out())

# display the extracted features
features_df
```

# MACHINE LEARNING ALGORITHM

```python
from sklearn.model_selection import train_test_split

# split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_tfidf,      # feature matrix
    y,            # target labels (sentiment)
    test_size=0.2,  # 20% of the data will be used for testing
    random_state=42 # ensures reproducibility
)

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)
```

# ACCURACY TESTING SET

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report

# create the MLP model
mlp_model = MLPClassifier(
    hidden_layer_sizes=(512, 256, 128),  # three hidden layers with 512, 256, and 128 neurons
    activation='relu',                  # ReLU activation function
    solver='adam',                      # adam optimizer
    alpha=1.0,                          # L2 regularization term
    learning_rate='adaptive',          # adaptive learning rate
    max_iter=1000,                      # maximum number of iterations
    early_stopping=True,               # stop early if validation score doesn't improve
    n_iter_no_change=10,               # stop after 10 epochs with no improvement
    random_state=42                    # seed for reproducibility
)

# train the model
mlp_model.fit(X_train, y_train_enc)

# predict on training and testing sets
y_pred_train_mlp = mlp_model.predict(X_train)
y_pred_test_mlp = mlp_model.predict(X_test)

# evaluate accuracy
accuracy_train_mlp = accuracy_score(y_train_enc, y_pred_train_mlp)
accuracy_test_mlp = accuracy_score(y_test_enc, y_pred_test_mlp)

# display evaluation results
print(f'MLP Training Accuracy: {accuracy_train_mlp:.4f}')
print(f'MLP Testing Accuracy : {accuracy_test_mlp:.4f}')
print('\nClassification Report (Testing - MLP):\n', classification_report(y_test_enc, y_pred_test_mlp))
```
